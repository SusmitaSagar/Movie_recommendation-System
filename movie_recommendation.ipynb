```python
#change all the above keywords in text from key:value pair
import ast
def convert(obj):
  L =[]
  for i in ast.literal_eval(obj): #we get one list from string of list  ast.literal_eval for converting into list
      L.append(i['name'])
  return L

```

```python
#change all the above keywords in text from key:value pair
import ast
def convert3(obj):
  L =[]
  counter = 0
  for i in ast.literal_eval(obj): #we get one list from string of list  ast.literal_eval for converting into list
      if counter != 3:
         L.append(i['name'])
         counter+=1
      else:
        break
  return L
```

```python
 #change all the above keywords in text from key:value pair
import ast
def fetch_director(obj):

  L =[]
  for i in ast.literal_eval(obj): #we get one list from string of list  ast.literal_eval for converting into list
      if i['job'] == 'Director':

         L.append(i['name'])
  return L
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
#make a helper fn
def stem(text):
  y = []
  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)

```

```python
#the order of the array will chnage if we sort it, thus we need to hold the index value o all the vectors.
#use enumerate fn for that
#sorted(list(enumerate(similarity[0])),reverse=True, key=lambda x:x[1])[1:6]
```

```python
#steps of this fn
#go to the index of given movie and then find its similarity vector.sort the vectorand get top 5 vectors and get their index and title name

def recommend(movie):
  movie_index = new_df[new_df['title'] == movie].index[0]
  distances = similarity[movie_index]
  movies_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
  for i in movies_list:
     print(new_df.iloc[i[0]].title)


```

```python
recommend('Avatar')
```

```python
recommend('Newlyweds')
```

```python
recommend('Batman Begins')
```
